```python
import numpy as np
import pandas as pd
```

```python
pd.DataFrame(np.load('mp_cnn_2/val_metrics.npy', allow_pickle=True).tolist()[1])
```

```python
training_metrics_list  = np.load('mp_cnn_2/training_metrics.npy', allow_pickle=True).tolist()
val_metrics_list  = np.load('mp_cnn_2/val_metrics.npy', allow_pickle=True).tolist()
training_dfs=[]
val_dfs=[]

for i in training_metrics_list:
    training_dfs.append(pd.DataFrame(i,columns=['RMSE', 'MSE', 'MAE', 'MAPE', 'R2']))
    
for i in val_metrics_list:
    val_dfs.append(pd.DataFrame(i,columns=['RMSE', 'MSE', 'MAE', 'MAPE', 'R2']))    
```

```python
param_list = []
for filters in [3,6,9,12,15]:
    for kernel_size in [3,5,7,9,11,13,15]:
        param_list.append(str(filters) + 'filters, ' +str(kernel_size) +'kernel_size')
        
```

```python
training_dict = dict(zip(param_list, training_dfs))
val_dict = dict(zip(param_list, val_dfs))
```

```python
param_list
```

```python
val_dict_avg=[]
for key in val_dict:
    val_dict_avg.append((val_dict[key].mean()))
```
